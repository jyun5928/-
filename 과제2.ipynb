```python
#K-means 클러스터링 
import matplotlib.pyplot as plt

# 'Age'와 'Annual Income (k$)' 컬럼을 사용
X = df[['Age', 'Annual Income (k$)']]

# K-means 클러스터링 (군집 수 3으로 설정)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# 결과 시각화
plt.scatter(df['Age'], df['Annual Income (k$)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('K-means Clustering')
plt.show()
```

```python
# 특성값 평균
kmeans.cluster_centers_
```

```python
# 응집도 확인
kmeans.inertia_
```

```python
# 응집도 시각화
from sklearn.cluster import KMeans
cs = []
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 5), cs)
plt.title('Mall_Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# 계층적 군집화 (Hierarchical Clustering)
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns

# 'Age'와 'Annual Income (k$)' 컬럼 사용
X = df[['Age', 'Annual Income (k$)']]

# 계층적 군집화
clustering = AgglomerativeClustering(n_clusters=3)
df['Cluster'] = clustering.fit_predict(X)

# 결과 시각화
sns.scatterplot(x='Age', y='Annual Income (k$)', hue='Cluster', data=df, palette='viridis')
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# 'Age'와 'Annual Income (k$)' 컬럼 사용
X = df[['Age', 'Annual Income (k$)']]

# DBSCAN 클러스터링
dbscan = DBSCAN(eps=3, min_samples=5)
df['Cluster'] = dbscan.fit_predict(X)

# 결과 시각화
plt.scatter(df['Age'], df['Annual Income (k$)'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# 엘보우
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 'Age'와 'Annual Income (k$)' 컬럼 사용
X = df[['Age', 'Annual Income (k$)']]

# WSS 값 계산
wss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wss.append(kmeans.inertia_)

# 엘보우 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WSS (Within-cluster Sum of Squares)')
plt.show()
```

```python
# 실루엣
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 'Age'와 'Annual Income (k$)' 컬럼 사용
X = df[['Age', 'Annual Income (k$)']]

# 실루엣 점수 계산
sil_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    sil_scores.append(score)

# 실루엣 점수 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), sil_scores, marker='o')
plt.title('Silhouette Scores for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# 최적의 클러스터 수
optimal_k = range(2, 11)[sil_scores.index(max(sil_scores))]
print(f"Optimal number of clusters based on silhouette score: {optimal_k}")
```

```python
# 2D 시각화
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# 'Age'와 'Annual Income (k$)' 컬럼 사용
X = df[['Age', 'Annual Income (k$)']]

# KMeans 클러스터링
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# 클러스터링 결과 2D 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Annual Income (k$)', data=df, hue='Cluster', palette='Set1', s=100, alpha=0.7)

# 시각화 제목 & 레이블 추가
plt.title('Customer Segmentation based on Age and Annual Income', fontsize=15)
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend(title='Cluster')

# 플롯을 출력
plt.show()
```

```python
# 3D 시각화
from mpl_toolkits.mplot3d import Axes3D

# 'Age', 'Annual Income (k$)', 'Spending Score (1-100)' 컬럼 사용
X_3d = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# KMeans 클러스터링
kmeans_3d = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans_3d.fit_predict(X_3d)

# 3D 시각화
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# x, y, z 값 설정
x = df['Age']
y = df['Annual Income (k$)']
z = df['Spending Score (1-100)']

# 3D 산점도
scatter = ax.scatter(x, y, z, c=df['Cluster'], cmap='Set1', s=100, alpha=0.7)

# 제목 및 레이블 설정
ax.set_title('3D Customer Segmentation', fontsize=15)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')

# 컬러 바 표시
plt.colorbar(scatter)

plt.show()
```
